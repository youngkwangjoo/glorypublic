##본격적인 행맨 게임 규칙 설명

- words 리스트에 추가된 단어들을 사용해주세요
  - words = ["apple", "banana", "orange", "grape", "lemon"]

- 단어의 길이에 맞게 밑줄을 출력해주세요  
  (예)banana의 경우 _ _ _ _ _ _

- 사용자부터 1글자씩 입력을 받되,
   - 단어에 입력값이 포함되면 "단어를 맞추셨군요. 더 힘내봐요"라고 출력해주세요"
   - 단어에 입력값이 포함되어 있지 않다면 "틀렸습니다. 남은 시도 횟수 "란 메시지와 함꼐 남은 횟수를 출력해주세요

- 매번 입력을 받을 때마다 현재까지 맞힌 글자들 표시(맞히면 글자는 밑줄 출력)
    - 예) a 입력 시 : a _ _ _ _
          p 입력 시 : a p p _ _
          c 입력 시 : a p p _ _

- 모든 단어를 맞췄을 경우 "축하합니다! 단어를 맞추셨습니다"와 시도 횟수를 출력해주세요


행맨은 정말 유명 게임으로 찾아볼 수 있는 자료들이 많습니다.   
문제가 어려우신 경우 구글링 통해 문제를 해결해보세요.  

```python
#단어 길이맞게 표시하기 "_"를 len만큼 표시
#str(input) 으로 알파벳 입력을 받고 같은 문자열이나오면 true print(a____?), 입력 false가 나오면 else로 넘겨서 print(틀렸습니다.)출력

#문제 난이도가 넘 높아서 정답 코드 해석을 했습니다. 
import random

def choose_word():
    words = ['apple', 'banana', 'orange', 'grape', 'lemon']
    return random.choice(words)
# 풀 문제를 랜덤으로 호출함
def display_word(word, guessed_letters):
    display = ''
    for letter in word:
        if letter in guessed_letters:
            display += letter
        else:
            display += '_'
    return display
#행맨의 기본 규칙설정
# 매개변수를 word, guessed_letters 두개 받고, 변수 display는 빈칸, letter이라는 단어가 2번 매개변수에 있으면 변수는 letter씩 증가한다. 
# 그게 아니면 _가 하나씩 늘어나는걸 반복한다. 

def hangman():
    word = choose_word()
    #단어를 랜덤 호출
    guessed_letters = []
    #2번 매개변수를 리스트 빈값으로 설정
    attempts_left = 6
    #시도횟수는 6번으로 고정
    #기본 변수선언
    print("행맨 게임을 시작합니다.")
    print(display_word(word, guessed_letters))
    #2번 함수를 호출
    while True:
        guess = input("알파벳을 입력하세요: ").lower()
        #입력값을 소문자로 받음
        if guess in guessed_letters:
            print("중복된 글자를 쓰셨습니다!")
            continue
        #중복 입력시 출력문구
        guessed_letters.append(guess)
        #입력값이 2번 매개변수에 있을 경우 프린트 출력이후 계속진행, 입력값 추가
        if guess not in word:
            attempts_left -= 1
            print(f"그런 알파벳은 들어가있지 않아요: {attempts_left}")
            if attempts_left == 0:
                print("땡! 정답 단어는", word, "입니다")
                break
        else:
            print("정답입니다. ")
        #입력값이 없을 경우 횟수 차감, 횟수가 0이 되면 print출력 이후 강제종료 break
        word_display = display_word(word, guessed_letters)
        print(word_display)
        #이후 정답공개
        if '_' not in word_display:
            print("단어를 맞추셨습니다.")
            break
        #_이 없어진다는건 정답을 맞춘거니까 맞추면 강제 종료
hangman( )
```
